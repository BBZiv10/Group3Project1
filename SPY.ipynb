```python
import yfinance as yf  
import matplotlib.pyplot as plt
import pandas as pd 
import pathlib as Path

%matplotlib inline 
```

```python
data_spy = yf.download('SPY','2021-01-01','2021-06-15')
data_spy["Adj Close"].plot()
plt.show()
```

```python
data_btc = yf.download('BTC-USD','2021-01-01','2021-06-15')
data_btc["Adj Close"].plot()
plt.show()
```

```python
up = data_spy['Adj Close'].pct_change()
up.plot() 
```

```python
up1 = data_btc['Adj Close'].pct_change()
up1.plot(color= 'green', kind='bar')
```

```python
up1.plot(kind='bar')
```

```python
up.plot(kind='bar', color='green', figsize=(12.5, 4.8))
up1.plot(kind='bar', color='red', label='SPY & BTC Daily % Changes')
```

```python
#Concate percentage change data of SPY and BTC into a DF

data = pd.DataFrame()

data['SPY up 1%'] = up.pct_change() > .01
data['SPY down 1%'] = up.pct_change() < -.01
data['BTC up 5%'] = up1.pct_change() > .05
data['BTC down 5%'] = up1.pct_change() < -.05
```

```python
#Export DF to a csv 

data.to_csv(r'C:\Users\Luke\Desktop\Project 1\Group3Project1\pct_change.csv')
```
